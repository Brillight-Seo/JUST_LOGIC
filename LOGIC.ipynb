```python
def print_board(board,N):
    for i in range(N):
        for j in range(N):
            if board[i][j]==0:
                print('□',end='')
            elif board[i][j]==-1:
                print('※',end='')
            else:
                print('■',end='')
        print()
    print()
```

```python
def roate_board(board,N):
    temp_board=[[0 for _ in range(N)] for _ in range(N)]
    for i in range(N):
        for j in range(N):
            temp_board[i][j]=board[j][i]
    return temp_board;
```

```python
import copy
def func1(inputted, len1, hint):
    temp=copy.deepcopy(inputted)
    if len(hint)==1:
        if hint[0]==0:
            return [-1 for _ in range(len1)]
        count=0
        for i in range(len1):
            if i+hint[0] <=len1:
                count+=1
                for j in range(i,i++hint[0]):
                    temp[j]+=1
        ttemp=[0 for _ in range(len1)]
        for t in range(len(temp)):
            if temp[t] ==count:
                ttemp[t]=1;
            else: # 기존거 유지하도록 변경
                ttemp[t]=inputted[t];
        return ttemp;
    else:
        return temp;
            
```
